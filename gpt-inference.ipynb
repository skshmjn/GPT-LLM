```python
from torch import cuda
from huggingface_hub import login
import transformers
```

```python
device = f'cuda:{cuda.current_device()}' if cuda.is_available() else 'cpu'
device
```

```python
model_id = "meta-llama/Llama-3.2-1B"
```

```python
access_token_read = "<YOUR HUGGINGFACE Token>"
```

```python
login(token = access_token_read)
```

```python
tokenizer = transformers.AutoTokenizer.from_pretrained(model_id)
```

```python
generate_text = transformers.pipeline(
    model=model_id,
    tokenizer=tokenizer,
    task='text-generation',
    temperature=0.0001,  # 'randomness' of outputs, 0.0 is the min and 1.0 the max
    max_new_tokens=512,  # mex number of tokens to generate in the output
    repetition_penalty=4.1,  # without this output begins repeating
    device_map = 'auto'
)
```

```python
%%time
res = generate_text("Explain to me the difference between star and planets.")
print(res[0]["generated_text"])
```
